```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('download.jpeg')

kernel = np.ones((5, 5), np.float32)/25
dst = cv2.filter2D(img, -3, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(dst), plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
blue = cv2.blur(img, (8, 1))
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.subplot(122), plt.imshow(blue), plt.title('Blur image')
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('blur.jpeg')
median = cv2.medianBlur(image,15)
plt.subplot(121), plt.imshow(image), plt.title('Original')
plt.subplot(122), plt.imshow(median), plt.title('Median blur image')
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
